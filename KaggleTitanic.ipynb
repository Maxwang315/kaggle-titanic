### Handling Empty / String Data

```python
#See all the null values
print(train_data.isnull().sum())
```

```python
#We don't need to worry about cabin, since we are not using it in X and y. For age, we will fill the null values with the mean, 
#and with mode for embarked
train_data['Age'] = train_data['Age'].fillna(train_data['Age'].median()) 
train_data['Embarked'] = train_data['Embarked'].fillna(train_data['Embarked'].mode()[0])
print(train_data.isnull().sum())
```

```python
usedFeatures = ['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked']
X = train_data[usedFeatures]
y = train_data['Survived']
```

```python
#One hot encode the String types

X = pd.get_dummies(X, columns = ['Sex', 'Embarked'])
```
